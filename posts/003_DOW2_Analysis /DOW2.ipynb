# The Different Racial Groups

![](img/countries.png){width=400}

### `race`was chosen from a drop down menu in the survey, in which it asked you to identify your race and write its associated number. The different racial groups represented were as followed which I could reference for my analysis: ```
1=**Mixed Race**, 2=**Arctic (Siberian, Eskimo)**, 3=**Caucasian (European)**, 4=**Caucasian (Indian)**, 5=**Caucasian (Middle East)**, 6=**Caucasian (North African, Other)**, 7=**Indigenous Australian**, 8=**Native American**, 9=vNorth East Asian (Mongol, Tibetan, Korean Japanese, etc)**, 10=**Pacific (Polynesian, Micronesian, etc)**, 11=**South East Asian (Chinese, Thai, Malay, Filipino, etc)**, 12=**West African, Bushmen, Ethiopian** 13=**Other** (0=missed)
```

````python
race_values = '''1=Mixed Race, 2=Arctic (Siberian, Eskimo), 3=Caucasian (European), 
4=Caucasian (Indian), 5=Caucasian (Middle East), 6=Caucasian (North African, Other), 
7=Indigenous Australian, 8=Native American, 
9=North East Asian (Mongol, Tibetan, Korean Japanese, etc), 
10=Pacific (Polynesian, Micronesian, etc), 
11=South East Asian (Chinese, Thai, Malay, Filipino, etc), 12=West African, Bushmen, Ethiopian, 13=Other, 0=missed
'''

parts = re.split(r'(?:,\s+)?([0-9]+)=', race_values.strip())

race_keys = [int(key_val) for key_val in parts[1::2]]
race_values = parts[2::2]
````

````python
race_mapping = dict(zip(race_keys,race_values))
race_mapping
````

````python
big5_df=big5_df.assign(race_cat=big5_df['race'].map(race_mapping))
````

## Defining extroversion in this big data set: 

To actually start finding the answer to the question, I have to find the extroversion score for each user. To find extroversion from the entire test, I summed the responses (numerical values 1-being strong no, 2-being strong yes) of these **five questions** in the 49 question set that described extroverted qualities in the best capacity:
- E1: "I am the life of the party"
- E3: "I feel comfortable around peoplea
- E5: "I start conversations"
- E7: "I talk to a lot of different people at parties"
- E9: "I don't mind being the center of attention" 

#### Each numerical value for the questions were added to make a total that defined each user's Extroversion Value:

````python
big5_df["E_value"] = big5_df[["E1", "E3", "E5", "E7", "E9" ]].sum(axis=1)
````

### This would create a seperate column in the compiled chart that gave each user an Extroversion score or E_value

### It looked something like this:

````python
big5_df.head(5)
````

### Now I had to find the important stuff which was **association**. This was found by grouping each user by race, and then averaging the Extroversion scores (E_value) to calculate the total average extroversion score for each race: 

````python
group = big5_df.groupby("race_cat") 
````

````python
race_avg = group["E_value"].mean()
````
