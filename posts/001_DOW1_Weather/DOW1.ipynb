## 1) What was the weather like in Philadelphia the day of my birthday (01-30) for each year since 2006-2025?
  ## * a. What was the average fahrenheit for the day?
  ## * b. Did it rain?
  ## * c. Did it snow?

```python
weather_df.query('day =="January 30"')
```

## A) What was the average fahrenheit for the day? 

```python
birthday = weather_df.query('day == "January 30"') 
temp_avg = (birthday['high'] + birthday['low'])/2
weather_df = weather_df.assign(temp_avg=(weather_df['high']+weather_df['low'])/2)
```

```python
weather_df.query('day =="January 30"')
```

```python
birthday_mean_temp = temp_avg.mean()
```

```python
birthday_mean_temp
```

#### Since I was born in 2006, combining the weather data of Philadelphia for the past twenty years, the average fahrenheit on my birthday in Philadelphia is 32.3 degrees. 

## B) Did it rain? 

```python
birthday = weather_df.query('day == "January 30"') 
is_rain = birthday['rain']>0
```

```python
is_rain
```

```python
is_rain.value_counts()
```

#### Since I was born in 2006, majority of the years since I was alive on the day of my birthday in Philadelphia it did not rain. (13 years it did not and 7 years it did). 

## C) Did it snow? 

```python
birthday = weather_df.query('day == "January 30"') 
is_snow = birthday['snow']>0
```

```python
is_snow.value_counts()
```

#### Since I was born in 2006, majority of the years since I was alive on the day of my birthday in Philadelphia it did not snow. (14 years it did not and 6 years it did).
